### 📦 Step 1: Install Python Dependencies
Installs all necessary packages including Gradio, Diffusers, Safetensors, and OpenCV.

```python
import subprocess, sys

packages = [
    "numpy>=1.24.0,<2.0.0",
    "gradio>=4.40.0",
    "diffusers>=0.30.0",
    "transformers>=4.46.3",
    "accelerate>=0.26.0",
    "peft>=0.8.0",
    "controlnet-aux>=0.0.10",
    "rembg>=2.0.68",
    "onnxruntime>=1.16.0",
    "opencv-python-headless>=4.8.0",
    "scipy>=1.11.0",
    "scikit-image>=0.21.0",
    "huggingface_hub>=0.20.0",
    "safetensors>=0.4.0",
    "xformers"
]

print("📥 Starting pip installation (approx 2 mins)...")
subprocess.check_call([sys.executable, "-m", "pip", "install", *packages, "-q"])
print("✅ All dependencies installed successfully!")
```

### 🧬 Step 2: Clone/Sync SnapStudio AI Repository
Downloads/pulls the application workspace code to the Kaggle notebook.

```python
import os, subprocess, sys

REPO_URL = "https://github.com/chaitanyabhujbal912006-afk/Snapstudio_AI.git"
REPO_DIR = "/kaggle/working/snapstudio"

if not os.path.exists(REPO_DIR):
    print(f"📥 Cloning repository from {REPO_URL}...")
    subprocess.check_call(["git", "clone", REPO_URL, REPO_DIR, "--depth=1"])
    print("✅ Repository cloned!")
else:
    print("🔄 Pulling latest updates...")
    subprocess.check_call(["git", "-C", REPO_DIR, "pull"])
    print("✅ Repository updated!")

os.chdir(REPO_DIR)
sys.path.insert(0, REPO_DIR)
print(f"📍 Working Directory: {os.getcwd()}")
```

### ⚙️ Step 3: Check GPU Hardware and VRAM Size
Ensures PyTorch registers the active T4 GPU.

```python
import torch

if torch.cuda.is_available():
    print(f"✅ T4 GPU Detected: {torch.cuda.get_device_name(0)}")
    print(f"   VRAM Profile: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB Available")
else:
    print("❌ WARNING: No GPU detected. The notebook will run extremely slowly.")
    print("   Please verify that Accelerator is set to 'GPU T4 x2' in Kaggle Notebook settings.")
```
